```python
#http://firsttimeprogrammer.blogspot.com/2015/04/fourier-series-and-square-wave.html
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
T = 8

def squareWave(x):
    global T
    lowerBoundLeft = (-T/2)
    lowerBoundRight = 0
    upperBoundLeft = 0
    upperBoundRight = (T/2)
    one = 1
    negativeOne = -1
    while True:
        if (x >= lowerBoundLeft) and (x <= lowerBoundRight):
            return negativeOne
        elif (x >= upperBoundLeft) and (x <= upperBoundRight):
            return one
        else:
            lowerBoundLeft -= T/2
            lowerBoundRight -= T/2
            upperBoundLeft += T/2
            upperBoundRight += T/2
            if one == 1:
                one = -1
                negativeOne = 1
            else:
                one = 1
                negativeOne = -1

# Coeficientes Bn 
def bn(n):
    n = int(n)
    if (n%2 != 0):
        return 4/(np.pi*n)
    else:
        return 0

# Wn
def wn(n):
    global T
    wn = (2*np.pi*n)/T
    return wn

# Series de Fourier 
def fourierSeries(n_max,x):
    a0 = 0
    partialSums = a0
    for n in range(1,n_max):
        try:
            partialSums = partialSums + bn(n)*np.sin(wn(n)*x)
        except:
            print("pass")
            pass
    return partialSums

def grafica(x_,y,f,armonics):
    plt.figure(figsize=(16,10))
    plt.plot(x_,y,color="blue",label="SeÃ±al")
    plt.plot(x_,f,color="red",label="Aproximacion de Fourier")
    plt.title("Series de Fourier, aproximacion con "+str(armonics)+" numero de armonicos")
    plt.legend(loc=4)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.show()
    
```

```python
# Setup
x_ = np.linspace(-20,20,10000)
armonics = 2
y = []
f = []
for i in x_:
    y.append(squareWave(i))
    f.append(fourierSeries(armonics,i))
grafica(x_,y,f,armonics)
```
