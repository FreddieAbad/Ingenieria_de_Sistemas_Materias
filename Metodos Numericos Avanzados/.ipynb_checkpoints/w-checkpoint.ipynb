# Metodo del gradiente - máximo descenso

https://github.com/vlarobbyk/descenso-por-gradiente/blob/master/GradientDescent-1.ipynb

https://gist.github.com/dirmeier/d9f53587e32f265d92b1039d5d9a168f

```julia
using Gadfly
using Distributions
using Colors

function df(x, y, b)
    sum(- (y - x*b)' * x)
end

function metodogradiente()
   rnorm = Normal()
   x = rand(rnorm, 100)
   y = x * 2 + rand(rnorm, 100)
   b    = 0.0
   bold = 1.0
   bvals = Float64[]
   while (sum(abs(b - bold)) > 1e-10)
       push!(bvals, b)
       bold = b
       der = df(x, y, bold)
       b   =  b - 0.001 * der
   end
   plot(Guide.title("Valor Real: 2, Valor Aproximado: " * string(b)),
        layer(x=x, y=y, Geom.point()),
        layer(x=x,y=x*bvals[1], Geom.line()),
        layer(x=x,y=x*bvals[5], Geom.line()),
        layer(x=x,y=x*b, Geom.line()))
end
```

```julia
metodogradiente()
```

```julia
#Tiempo de Ejecucion
@time metodogradiente();
```

# Metodo de gradientes conjugados

https://groups.google.com/forum/#!topic/julia-users/jsAA7M7L68E


https://github.com/JuliaNLSolvers/Optim.jl/blob/master/docs/src/algo/cg.md

```julia
f(x) = (1.0 - x[1])^2 + 100.0 * (x[2] - x[1]^2)^2
function g!(storage, x)
    storage[1] = -2.0 * (1.0 - x[1]) - 400.0 * (x[2] - x[1]^2) * x[1]
    storage[2] = 200.0 * (x[2] - x[1]^2)
end
```

```julia
optimize(f, g!, zeros(2), ConjugateGradient())
```

```julia
@time optimize(f, g!, zeros(2), ConjugateGradient());
```
